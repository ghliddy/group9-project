```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
goa_data = "https://raw.githubusercontent.com/ghliddy/group9-project/main/goaon-asset_list-20230326T102704.csv"

csv = goa_data + "https://raw.githubusercontent.com/ghliddy/group9-project/main/goaon-asset_list-20230326T102704.csv"

df = pd.read_csv(goa_data)

df.head()
```

```python
pd.read_csv(goa_data, header=None, skiprows=0)
```

```python
df.columns
df.dtypes


```

```python
new_lat = []

for lat in df['Lat']:
  parts = lat.split()
  first_part = parts[0]
  new_lat.append(first_part)
```

```python
df['Lat'] = new_lat
```

```python
df['Lat'] = df['Lat'].astype(int)
```

```python
df['Lat']
```

```python
new_lon = []

for lon in df['Lon']:
  parts = lon.split()
  first_part = parts[0]
  new_lon.append(first_part)


df['Lon'] = new_lon
df['Lon']
```

```python
df.to_csv('new.csv')
```

The code above was used to trim the values for latitude and longitude. By doing this, we a triming string variables from the values so that the data can be displayed in arcGIS. Below is an image of the map that was produced from this data.

The column data was cleaned to express only the standards needed for map reading. The point data displays not only the specific location of the research device (ship, Argo float...), but also the institute/country that deployed the device, the body of water where the device is deployed, and the measurements it is taking. Each device is researching different objectives based on its location. For example, the devices deployed in the Mediterranean Sea are measuring salinity as a major component. Rates of salinity in the Med. Sea are higher as a result of decreased current circulation and age of the sea bed. Research devices that are observing data in areas of major upwelling and current circulation (Gulf Stream, North Atlantic Current) are observing factors such as Zooplankton and Chlorophyll.
